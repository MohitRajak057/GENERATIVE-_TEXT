## Step 2 — Import Libraries

```python
import torch
import textwrap
from transformers import GPT2LMHeadModel, GPT2Tokenizer

print("✅ Libraries imported successfully!")
print(f"PyTorch version: {torch.__version__}")
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")
```

## Step 3 — Load Pre-trained GPT-2 Model

```python
print("⏳ Loading GPT-2 model and tokenizer...")

MODEL_NAME = "gpt2"  # GPT-2 small (117M parameters) — fast & lightweight

tokenizer = GPT2Tokenizer.from_pretrained(MODEL_NAME)
tokenizer.pad_token = tokenizer.eos_token  # Set padding token

model = GPT2LMHeadModel.from_pretrained(MODEL_NAME)
model = model.to(device)
model.eval()  # Set to evaluation mode

print("✅ GPT-2 model loaded successfully!")
print(f"Model parameters: ~117 Million")
```

## Step 4 — Define Text Generation Function

```python
def generate_text(prompt, max_length=200, num_sequences=1,
                  temperature=0.85, top_k=50, top_p=0.92):
    """
    Generate coherent text based on a given prompt using GPT-2.

    Parameters:
    -----------
    prompt       : str   - The input text prompt
    max_length   : int   - Maximum tokens to generate
    num_sequences: int   - Number of different outputs to generate
    temperature  : float - Controls randomness (lower = more focused)
    top_k        : int   - Top-K sampling (diversity control)
    top_p        : float - Nucleus sampling threshold

    Returns:
    --------
    list of generated text strings
    """
    inputs = tokenizer(
        prompt,
        return_tensors='pt',
        padding=True
    ).to(device)

    with torch.no_grad():
        output_ids = model.generate(
            inputs['input_ids'],
            attention_mask=inputs['attention_mask'],
            max_length=max_length,
            num_return_sequences=num_sequences,
            do_sample=True,
            temperature=temperature,
            top_k=top_k,
            top_p=top_p,
            pad_token_id=tokenizer.eos_token_id,
            no_repeat_ngram_size=3,      # Avoid repetitive phrases
            early_stopping=False
        )

    results = []
    for ids in output_ids:
        text = tokenizer.decode(ids, skip_special_tokens=True)
        results.append(text)

    return results


def display_generated_text(prompt, generated_texts):
    """Nicely display the prompt and generated text."""
    print("=" * 65)
    print(f"📝 PROMPT: {prompt}")
    print("=" * 65)
    for i, text in enumerate(generated_texts, 1):
        print(f"\n🔹 Generated Output {i}:")
        print("-" * 50)
        # Wrap text at 70 characters for clean display
        wrapped = textwrap.fill(text, width=70)
        print(wrapped)
    print("\n")


print("✅ Text generation function defined!")
```

## Step 5 — Generate Text on Different Topics

Demonstrating GPT-2 text generation across **5 diverse topics**.

### 🧠 Topic 1: Artificial Intelligence

```python
prompt_ai = "Artificial intelligence is transforming the world by"

output_ai = generate_text(prompt_ai, max_length=180, num_sequences=2)
display_generated_text(prompt_ai, output_ai)
```

### 🌍 Topic 2: Climate Change

```python
prompt_climate = "Climate change is one of the most pressing challenges of our time because"

output_climate = generate_text(prompt_climate, max_length=180, num_sequences=2)
display_generated_text(prompt_climate, output_climate)
```

### 🚀 Topic 3: Space Exploration

## Step 6 — Interactive Text Generator

A function that allows generating text for **any custom user prompt**.

```python
def interactive_generate(user_prompt, length=150, variations=1):
    """
    Interactive wrapper — enter any topic prompt and get generated text.

    Parameters:
    -----------
    user_prompt  : str - Your custom prompt/topic starter
    length       : int - How long the output should be (tokens)
    variations   : int - How many different paragraphs to generate
    """
    print(f"🎯 Generating {variations} paragraph(s) for your prompt...\n")
    results = generate_text(user_prompt, max_length=length, num_sequences=variations)
    display_generated_text(user_prompt, results)


# ---- TRY YOUR OWN PROMPT HERE ----
my_prompt = "The future of renewable energy depends on"
interactive_generate(my_prompt, length=180, variations=2)
```

## Step 7 — Exploring Generation Parameters

Demonstrating how **temperature** affects creativity vs. coherence.

```python
test_prompt = "Machine learning algorithms help computers learn from data by"

temperature_values = [0.4, 0.85, 1.2]
labels = ["Low (0.4) — Focused & Predictable",
          "Medium (0.85) — Balanced",
          "High (1.2) — Creative & Random"]

print("🌡️ TEMPERATURE COMPARISON")
print("=" * 65)
print(f"Prompt: {test_prompt}\n")

for temp, label in zip(temperature_values, labels):
    result = generate_text(test_prompt, max_length=120,
                           num_sequences=1, temperature=temp)
    print(f"\n🔸 Temperature {label}:")
    print("-" * 50)
    print(textwrap.fill(result[0], width=70))

print("\n✅ Parameter comparison complete!")
```
